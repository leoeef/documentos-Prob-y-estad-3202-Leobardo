## Arrays

Un **array** es una estructura de datos que consiste en una colección de elementos (valores o variables), cada uno identificado por al menos un índice o clave. Un array se almacena de modo que la posición de cada elemento se pueda calcular a partir de su tupla de índice mediante una fórmula matemática. El tipo más simple de array es un array lineal, también llamado array unidimensional.

## Numpy

En numpy:
* Cada dimensión se denomina **axis**
* El número de dimensiones se denomina **rank**
* La lista de dimensiones con su correspondiente longitud se denomina **shape**
* El número total de elementos (multiplicación de la longitud de las dimensiones) se denomina **size**

## Acceso a los elementos de un Array
### Array Unidimensional 

```python
# Crear un Array unidimensional

# Crear un Array unidimensional a partir de una tupla
array_uni = np.array((1, 3, 5, 7, 9, 11))

# Imprime la forma (dimensiones) del array
print("shape: ", array_uni.shape)

# Imprime el contenido del array
print("Array uni: ", array_uni)
```

```python
# Accediendo al quinto elemento del array
array_uni[4]
```

```python
# Accediendo al tercer y cuarto elemento del Array
# Sintaxis: array[inicio : fin]
resultado = array_uni[2:4]

print(resultado)
```

**¿Cómo funciona el índice [2:4]?**

En Python y NumPy, el conteo de posiciones empieza en 0. Además, el rango funciona bajo la regla: "**el primero se incluye, el último se excluye**".

**1. El inicio (2):** Corresponde al tercer elemento.

**2. El fin (4):** Indica que el corte se detiene antes del índice 4.

```python
# Accediendo a elementos con un salto de 2
# Sintaxis: array[inicio : fin : salto]
resultado = array_uni[0::2]

print(resultado)
```

**¿Cómo funciona el índice [0::2]?**
La sintaxis de tres parámetros funciona así:

**Inicio (0):** Empezamos en el primer elemento (índice 0).

**Fin (vacío):** Al no poner nada entre los dos puntos, le decimos que llegue hasta el final del array.

**Salto (2):** Le indica que tome un elemento y se salte el siguiente. Es decir, va de 2 en 2.

### Array multidimencional 

```python
# Crear un array multidimensional (2 filas y 4 columnas)
# Se utilizan corchetes anidados: [[fila1], [fila2]]
array_multi = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])

# Imprime las dimensiones de la matriz
print("Shape:", array_multi.shape)

# Imprime la matriz completa
print("Array Multi:\n", array_multi)
```

```python
# Acceder al cuarto elemento de la primera fila
# Sintaxis: array[índice_fila, índice_columna]
elemento = array_multi[0, 3]

print(elemento)
```

```python
# Acceder a la segunda fila completa
# Sintaxis: array[índice_fila, todas_las_columnas]
segunda_fila = array_multi[1, :]

print(segunda_fila)
```

```python
# Acceder al tercer elemento de las 2 primeras filas
# Sintaxis: array[rango_filas, índice_columna]
resultado = array_multi[0:2, 2]

print(resultado)
```

```python
# Acceder al tercer elemento de las 2 primeras filas
# Sintaxis: array[rango_filas, índice_columna]
resultado = array_multi[0:2, 2]

print(resultado)
```

# Modificacion de un Array

## Crear un Array unidimencional e inicializarlo con el rango de elementos 0-27

```python
array1 = np.arange(28)
print("Shaper:", array1.shape)
print("Array 1:", array1)
```

```python
# Cambiar las dimenciones del Array y sus longitudes
array1.shape = (7,4)
print("Shaper:", array1.shape)
print("Array 1:\n", array1)
```

```python
# El ejemplo anterior devuelve un nuevo Array que apunta a los mismo datos
# Nota: Modificaciones en un Array, modifica el otro Array.
array2=array1.reshape(4,7)
print("Shaper:", array2.shape)
print("Array 2:\n", array2)
```

```python
# Modificar el nuevo Array devuelto
array2[0,3]=200
print("Array 2:\n", array2)
```

```python
print("Array 1:\n", array1)
```

```python
# Redimencionar a la dimecion original
# Nota: El nuevo arreglo apunta a los mismos datos.
print("Array 1:", array1.ravel())
```

## Operaciones Aitmeticas con Array

```python
# Crear dos arrays unidimencionales 
array1 = np.arange(2,18,2)
array2 = np.arange(8)
print("Array 1:\n", array1)
print("Array 2:\n", array2)
```

```python
# Suma
print(array1+array2)
```

```python
# Resta
print(array1-array2)
```

```python
# Multiplicacion
# Nota: No es una multiplicacion de matrices
print(array1*array2)
```
