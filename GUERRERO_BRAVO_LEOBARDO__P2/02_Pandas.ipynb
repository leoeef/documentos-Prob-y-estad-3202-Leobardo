# Introduccion a Pandas

[Pandas](https://pandas.pydata.org/) es una biblioteca que proporciona estructuras de datos y herramientas de analisis de datos de alto rendimiento y faciles de usar.
* La estructura de datos principal es el DataFrame, que puede considerarse como una tabla 2D en memoria (como una hoja de calculo, con nombre de columanas y etiquetas de fila).

## Estructura de Datos de Pandas

La biblioteca Pandas, de manera generica, contiene la siguiente estructura de datos:
* **Series**: Array de una dimension.
* **DataFrame**: Corresponde a una tabla de dos dimenciones.
* **Panel**: Similar a un Diccionario de DataFrame.

# Creacion de un Objeto Series

```python
# Crear un Objeto Series
s = pd.Series([2,4,6,8,10])
print(s)
```

```python
# Crear un objeto e inicializarlo con un diccionario de python
altura ={"Sherlyn": 160,"Daniel": 170, "Vannesa": 154, "Alejandro": 156}
s = pd.Series(altura)
print(s)
```

```python
# Crear un objeto series e inicializarlo con algunos de los elementos de un diccionario de python
altura ={"Sherlyn": 160,"Daniel": 170, "Vannesa": 154, "Alejandro": 156}
s = pd.Series(altura, index = ["Sherlyn", "Vannesa"])
print(s)
```

```python
# Crear un objeto series e inicializarlo con una scalar
s = pd.Series(34,["scalar1", "scalar2", "scalar3", "scalar4"])
print(s)
```

```python
# Accediendo al tercer elemento del Objeto
s["scalar3"]
```

## Acceso a los elementos de un elemento series

Cada elemento en un objeto series tiene un unico identificador que se denomina **indexlabel**.

```python
# Crea un objeto series
s = pd.Series([2,4,6,8], index=["num1", "num2", "num3", "num4"])
print(s)
```

```python
# Accedder al tercer elemento del objeto
s(2)
```

```python
# loc es la forma 
s.loc["num3"]
```

```python
# iloc es la forma estandar de acceder a un elemento de un objeto series por posiciom
s.iloc[2]
```

```python
# Acceder al segundo y tercer elmento de posicion 
s.iloc[2:4]
```

### Representancion grafica de un Objeto Series

# DataFrame
## Crear un objeto DataFrame

```python
# Crear un dataframe e inicializarlo con un diccionario de objetos series
Personas ={
    "Peso": pd.Series([50,70,58,51], ["Sherlyn", "Daniel", "Vannesa", "Alejandro"]),
    "Altura": pd.Series ({"Sherlyn": 160,"Daniel": 170, "Vannesa": 154, "Alejandro": 156}),
    "Hijos": pd.Series([2,1,], ["Sherlyn", "Vannesa"])
}
df = pd.DataFrame(Personas)
df
```

Puede forzaese al dataframe que presenta unas columnas determinadas y en orden determinado

```python
# Crear un DataFrame e inicializarlo con algun elemento  de un diccionario de Objetos Series
Personas ={
    "Peso": pd.Series([50,70,58,51], index= ["Sherlyn", "Daniel", "Vannesa", "Alejandro"]),
    "Altura": pd.Series ({"Sherlyn": 160,"Daniel": 170, "Vannesa": 154, "Alejandro": 156}),
    "Hijos": pd.Series([2,1,], index=["Sherlyn", "Vannesa"])
}
df = pd.DataFrame (
    data=Personas,
    columns = ["Altura","Peso"],
    index = ["Sherlyn", "Vannesa", "Alejandro"]
)
print(df)
```

### Evaluacion de expresiones sobre un DataFrame

```python
# Evaluar una funcion sobre una columna del DataFrame
df.eval("Altura/2")
```

```python
# Evaluar una funcion evaluando una variable local
max_altura = 170
df.eval("Altura >= @max_altura")
```

```python
# Aplicar una funcion externa del DataFrame
def func(x):
    return x + 2

df["Peso"].apply(func)
```

## Guardar y cargar un DataFrame

```python
Personas ={
    "Peso": pd.Series([50,70,58,51], index= ["Sherlyn", "Daniel", "Vannesa", "Alejandro"]),
    "Altura": pd.Series ({"Sherlyn": 160,"Daniel": 170, "Vannesa": 154, "Alejandro": 156}),
    "Hijos": pd.Series([2,1,], index=["Sherlyn", "Vannesa"])
}
df = pd.DataFrame (
    data=Personas,
    columns = ["Altura","Peso"],
    index = ["Sherlyn", "Vannesa", "Alejandro"]
)
print(df)
```

```python
# Guardar en distintos formatos
df.to_csv("Df_Personas.csv")
df.to_html("Df_Personas.html")
df.to_json("Df_Personas.json")
```
